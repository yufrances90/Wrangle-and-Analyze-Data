```python
import pandas as pd
import numpy as np
import os
import re
```

```python
data_dir = "../data"
```

```python
data_file_path = os.path.join(data_dir, "twitter-archive-enhanced.csv")
```

```python
df = pd.read_csv(data_file_path)
```

```python
df['text']
```

```python
df_new = df.copy()
```

```python
df_new.info()
```

```python
indices_with_retweets = df_new[df_new['retweeted_status_id'].notnull()].index
```

```python
df_new.drop(indices_with_retweets, inplace=True)
```

```python
for text in df_new['text']:
    print(text)
    print()
```

```python
p = re.compile('[0-9]+/10')
```

```python
bad_count = 0
none_count = 0
good_count = 0


for text, denominator, numerator in zip(df_new['text'].values, df_new['rating_denominator'].values, df_new['rating_numerator'].values):
    result = p.search(text)
    if result is not None and denominator != 10:
#         print(result.group(), numerator, denominator, "************************", text)
        bad_count += 1
    elif result is None:
        print(numerator, denominator, "************************", text)
        none_count += 1
    else:
        good_count += 1
        

print("**************************************************************************")        
print(good_count, none_count, bad_count, df_new.shape[0])
```

```python
text_to_remove = []

for text, denominator, numerator in zip(
    df_new['text'].values, 
    df_new['rating_denominator'].values, 
    df_new['rating_numerator'].values):
    
    result = p.search(text)
    
    if result is not None and denominator != 10:
        text_to_remove.append(text)
```
